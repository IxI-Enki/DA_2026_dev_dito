## J4: ContentAwareChunker flowchart

Flow reflects `pipeline/04_embeddings_creator/content_aware_chunker.py`: heading detection, paragraph/semantic split, size-based splitting, naive fixed-size fallback.

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis("off")

def box(ax, xy, w, h, text, style="round"):
    p = FancyBboxPatch(xy, w, h, boxstyle=style, facecolor="#E8EEF4", edgecolor="#2E4F8F", linewidth=1.5)
    ax.add_patch(p)
    ax.text(xy[0] + w/2, xy[1] + h/2, text, ha="center", va="center", fontsize=9)

def arrow(ax, start, end):
    ax.annotate("", xy=end, xytext=start, arrowprops=dict(arrowstyle="->", color="#2E4F8F", lw=1.5))

# Flow: Document -> should_skip? -> prepare_text -> get_config -> method -> (recursive_header | semantic | naive) -> Chunks
box(ax, (3.5, 8), 3, 0.6, "Document")
box(ax, (3.5, 6.8), 3, 0.6, "should_skip?")
box(ax, (3.5, 5.6), 3, 0.6, "prepare_text")
box(ax, (3.5, 4.4), 3, 0.6, "get_chunking_config")
box(ax, (3.5, 3.2), 3, 0.6, "method?")
box(ax, (0.5, 1.2), 2.2, 0.6, "recursive_header")
box(ax, (3.9, 1.2), 2.2, 0.6, "semantic")
box(ax, (7.3, 1.2), 2.2, 0.6, "naive")
box(ax, (3.5, 0.2), 3, 0.6, "Create Chunk objects")

arrow(ax, (5, 8), (5, 7.4))
arrow(ax, (5, 7.4), (5, 6.2))
arrow(ax, (5, 6.2), (5, 5.0))
arrow(ax, (5, 5.0), (5, 3.8))
arrow(ax, (5, 3.8), (5, 1.5))
arrow(ax, (1.6, 1.2), (3.5, 0.8))
arrow(ax, (5, 1.2), (5, 0.8))
arrow(ax, (8.4, 1.2), (6.5, 0.8))

ax.set_title("ContentAwareChunker flow (pipeline/04_embeddings_creator/content_aware_chunker.py)", fontweight="bold")
fig.tight_layout()
fig.savefig(FIGURES_DIR / "j4_content_aware_chunker_flowchart.png", bbox_inches="tight", dpi=DPI)
plt.close(fig)
print("Saved j4_content_aware_chunker_flowchart.png")
```
